```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(27,5))

df = pd.read_csv('2015.csv')
sns.boxplot(x='Region', y='Economy (GDP per Capita)', data=df)
# Ожидаемо высокие ВВП в Северной Америке,Европа и Австралии с малым рабросом.Самый большой разброс по ВВП замечен на Ближнем
# Востоке (есть очень богатые страны как ОАЭ, так крайне бедные, как Йемен).Африка ожиаемо на последнем месте по ВВП, причем
# самыебогатые страны беднее болшинства стран в мире.
```
