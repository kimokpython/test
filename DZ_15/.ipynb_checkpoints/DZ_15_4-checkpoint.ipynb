```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(22,5))

df = pd.read_csv('2015.csv')
df2 = pd.read_csv('2015-2019.csv')
df2 = df2[['Region','Happiness Score','Year']]
df2_grouped = df2.groupby(['Region','Year']).mean()
df2_grouped.head(25)
df2_pivot = df2_grouped.pivot_table(index='Year', columns='Region', values='Happiness Score', aggfunc=sum).fillna(0).applymap(float)
ax = sns.heatmap(df2_pivot)
# За 2015-2019 наблюдается крайнее слабое движение по среднему рейтингу счастья среди регионов. Самый стабильный регион - 
# Восточная Азия. Наименее стабильный регион - Латинская Америка.
```
