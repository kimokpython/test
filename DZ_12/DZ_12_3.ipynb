```python
import pandas as pd

def production_year(title):
    years = list(range(1950,2011))
    for year in years:        
        if str(year) in title:
            return year
            break
    return 1900    

movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv') 
df = pd.merge(movies, ratings, on='movieId', how='inner')
del df['userId']
del df['timestamp']
del df['movieId']

def groupby_mean(data):
    return data.mean()

df_grouped = df.groupby('title').apply(groupby_mean).reset_index()    
df_grouped['year'] = df_grouped['title'].apply(production_year)
df_years = df_grouped.groupby('year').apply(groupby_mean)
del df_years['year']
df_years.sort_values(by='rating', ascending = False).head(20)
```
