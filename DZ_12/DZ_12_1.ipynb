```python
import pandas as pd

movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv') 
df = pd.merge(movies, ratings, on='movieId', how='inner')
del df['userId']
del df['timestamp']

def groupby_mean(data):
    return data.mean()

df_grouped = df.groupby('title').apply(groupby_mean)

def movie_rating(rating):
    if rating > 4:
        return "High rating"
    elif rating > 2:
        return "Medium rating"
    else:
        return "Low rating"
    
df_grouped['class'] = df_grouped['rating'].apply(movie_rating)
df_grouped.head(20)
```
