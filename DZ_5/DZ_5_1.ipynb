```python
documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}
def main():
    command = ""
    while command != "q":
        command = input("Введите команду: ") 
        if command == "p": 
            print(owner(input("Введите номер документа: ")))         
        elif command == "s":  
            print(shelf(input("Введите номер документа: ")))
        elif command == "l":
            for document in documents:
                print("№:", document['number'], ", тип:", document['type'], ", владелец:", document['name'], ", полка хранения:", shelf(document['number'])[-1])  
        elif command == "as":
            print(add_shelf(input("Введите номер полки: ")))
        elif command == "ds":
            print(delete_shelf(input("Введите номер полки: ")))
        
def owner(number):    
    result = ""
    for document in documents:
        if document['number'] == number:
            result = "Владелец документа: " + document['name']
            break 
    if not result:
        result = "Документ не найден в базе"        
    return result
              
def shelf(number):    
    result = ""
    for key, value in directories.items():
        if number in value:
            result = "Документ хранится на полке: " + key
            break 
    if not result:
        result = "Документ не найден в базе"
    return result  
              
def add_shelf(number):    
    shelf_num = list(directories.keys())
    result = ','.join(shelf_num)
    if number in shelf_num:        
        result = "Такая полка уже существует. Текущий перечень полок: " + result
    else:
        directories[number] = ""
        result = "Полка добавлена. Текущий перечень полок: " + result + "," + number
    return result

def delete_shelf(number):
    shelf_num = list(directories.keys())
    result = ','.join(shelf_num)
    if number not in shelf_num:        
        result = "Такой полки не существует. Текущий перечень полок: " + result
    elif not directories[number]:
        directories[number] = ""
        del directories[number]
        result = "Полка удалена. Текущий перечень полок: " + result.replace(number,'')
        result = result.replace(",,",',')
    else:        
        result = "На полке есть документа, удалите их перед удалением полки. Текущий перечень полок: " + result        
    return result
    
main()     
```
